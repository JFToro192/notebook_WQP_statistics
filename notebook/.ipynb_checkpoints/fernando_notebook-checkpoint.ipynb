```python
url = "https://istsos.ddns.net/auth/realms/istsos/protocol/openid-connect/token"
token_payload = "username={}&password={}&grant_type=password&client_id=istsos-istsos".format(
    "satellite",
    "satellite"
)
req = requests.request(
    "POST", url,
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded'
    },
    data=token_payload
)
token_obj = req.json()
token = token_obj['access_token']
header = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {token}'
}

req_ds = requests.get(
    f"https://istsos.ddns.net/istsos/wa/istsos/services/demo/procedures/SATELLITE_CHL_TURB_CO_EAST",
    headers=header
)


ds = req_ds.json()['data']

print(ds)

req = requests.get(
    f"https://istsos.ddns.net/istsos/wa/istsos/services/demo/operations/getobservation/offerings/temporary/procedures/SATELLITE_CHL_TURB_CO_EAST/observedproperties/:/eventtime/last",
    headers=header
)
data = req.json()
data = data['data'][0]
data["observedProperty"]["component"] = list(filter(lambda x: "quality" not in x, data["observedProperty"]["component"]))
# print(data["result"])
data['result']["DataArray"]["field"] = list(filter(lambda x: "quality" not in x["name"], data['result']["DataArray"]["field"]))
# # print(
# #     list(filter(lambda x: "quality" not in x["name"], data['result']["DataArray"]["field"]))
# # )
# # break
# ds["assignedSensorId"]
values = [
    [
    "2020-01-01T00:01:00Z",
    2.86,
    1.21
    ],
    [
    "2020-01-02T00:01:00Z",
    3.42,
    1
    ],
    [
    "2020-01-05T00:01:00Z",
    3.11,
    1.28
    ],
    [
    "2020-01-08T00:01:00Z",
    3.31,
    1.12
    ],
    [
    "2020-01-11T00:01:00Z",
    3.16,
    1.13
    ],
    [
    "2020-01-12T00:01:00Z",
    3.5,
    0.84
    ],
    [
    "2020-01-16T00:01:00Z",
    3.89,
    0.91
    ],
    [
    "2020-01-17T00:01:00Z",
    3.48,
    1.03
    ]
]

data["samplingTime"] = {
    "beginPosition": "2020-01-01T00:01:00Z",
    "endPosition": "2020-01-17T00:01:00Z"
}
data['result']['DataArray']['values'] = values

res = requests.post(
    "https://istsos.ddns.net/istsos/wa/istsos/services/demo/operations/insertobservation",
    headers=header,
    data=json.dumps({
        "ForceInsert": "true",
        "AssignedSensorId": ds["assignedSensorId"],
        "Observation": data
    })
)
print(res.text)

# print(data)
```
